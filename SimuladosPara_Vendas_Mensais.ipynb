```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados simulados para vendas mensais, saldo financeiro e os 10 produtos mais vendidos
dados_mensais = pd.DataFrame({
    "Mês": ["Janeiro", "Fevereiro", "Março", "Abril"],
    "Vendas (R$)": [14270, 15050, 14580, 15970],
    "Saldo (R$)": [4138, 4970, 4020, 5540]
})

produtos_mais_vendidos = pd.DataFrame({
    "Produto": [
        "Toalha Banho Luxo", "Jogo de Lençol Queen", "Travesseiro Antialérgico",
        "Edredom Casal", "Toalha de Rosto", "Protetor de Colchão",
        "Pano de Prato Estampado", "Cortina Blackout", "Cobertor Casal",
        "Tapete de Banheiro"
    ],
    "Quantidade Vendida": [130, 120, 115, 105, 100, 95, 90, 85, 80, 75]
})

# Estoque inicial, entradas e saídas
estoque_dados = pd.DataFrame({
    "Produto": produtos_mais_vendidos["Produto"],
    "Estoque Inicial": [100]*10,
    "Entradas": [50, 60, 40, 50, 60, 30, 40, 50, 60, 50],
    "Saídas": produtos_mais_vendidos["Quantidade Vendida"]
})
estoque_dados["Estoque Final"] = estoque_dados["Estoque Inicial"] + estoque_dados["Entradas"] - estoque_dados["Saídas"]

# Gráfico 1: Vendas x Saldo
plt.figure(figsize=(8, 5))
sns.lineplot(data=dados_mensais, x="Mês", y="Vendas (R$)", marker="o", label="Vendas (R$)")
sns.lineplot(data=dados_mensais, x="Mês", y="Saldo (R$)", marker="s", label="Saldo (R$)")
plt.title("Evolução Mensal de Vendas e Saldo")
plt.ylabel("Valor (R$)")
plt.grid(True)
plt.legend()
plt.tight_layout()
grafico_vendas_saldo = "/mnt/data/grafico_vendas_saldo.png"
plt.savefig(grafico_vendas_saldo)
plt.close()

# Gráfico 2: Top 10 produtos mais vendidos
plt.figure(figsize=(8, 5))
sns.barplot(data=produtos_mais_vendidos, x="Quantidade Vendida", y="Produto", palette="viridis")
plt.title("Top 10 Produtos Mais Vendidos (Jan-Abr 2025)")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
plt.tight_layout()
grafico_top_produtos = "/mnt/data/top_10_produtos.png"
plt.savefig(grafico_top_produtos)
plt.close()

# Gráfico 3: Estoque final por produto
plt.figure(figsize=(8, 5))
estoque_plot = estoque_dados.sort_values("Estoque Final", ascending=False)
sns.barplot(data=estoque_plot, x="Estoque Final", y="Produto", palette="coolwarm")
plt.title("Estoque Final por Produto")
plt.xlabel("Unidades")
plt.tight_layout()
grafico_estoque = "/mnt/data/estoque_final_produtos.png"
plt.savefig(grafico_estoque)
plt.close()

grafico_vendas_saldo, grafico_top_produtos, grafico_estoque
```
