```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

# 1. Products DataFrame
products = pd.DataFrame([
    {"Código": f"{i:03d}", "Produto": p, "Custo (R$)": c, "Venda (R$)": v}
    for i, (p, c, v) in enumerate([
        ("Jogo de cama queen 200 fios", 80.00, 149.90),
        ("Jogo de cama casal floral", 75.00, 139.90),
        ("Edredom casal microfibra", 95.00, 169.90),
        ("Jogo de cama solteiro infantil", 65.00, 119.90),
        ("Jogo de lençol casal microfibra", 75.00, 139.90),
        ("Protetor de colchão impermeável", 40.00, 79.90),
        ("Fronha avulsa 100% algodão", 9.00, 18.90),
        ("Capa de almofada decorativa", 12.00, 24.90),
        ("Colcha matelassê queen", 70.00, 129.90),
        ("Travesseiro antialérgico", 25.00, 49.90),
        ("Kit travesseiro + fronha", 30.00, 59.90),
        ("Cobre-leito com porta-travesseiros", 85.00, 159.90),
        ("Toalha de banho luxo", 25.00, 49.90),
        ("Toalha de rosto bordada", 12.00, 22.90),
        ("Toalha de praia estampada", 30.00, 59.90),
        ("Jogo de toalhas 5 peças", 85.00, 159.90),
        ("Toalha infantil com capuz", 28.00, 54.90),
        ("Roupão felpudo unissex", 55.00, 99.90),
        ("Tapete para banheiro antiderrapante", 18.00, 34.90),
        ("Conjunto tapetes banheiro 3 peças", 40.00, 74.90),
        ("Toalha para academia (compacta)", 10.00, 19.90),
        ("Kit banheiro (toalhas + tapete)", 60.00, 114.90),
        ("Caminho de mesa rendado", 18.00, 34.90),
        ("Jogo americano floral (4 peças)", 14.00, 26.90),
        ("Jogo de guardanapos bordados", 16.00, 32.90),
        ("Toalha de mesa redonda estampada", 35.00, 64.90),
        ("Trilho de mesa natalino", 20.00, 39.90),
        ("Toalha de mesa retangular 6 lugares", 45.00, 84.90),
        ("Porta-copos em tecido (6 unid.)", 8.00, 15.90),
        ("Kit mesa posta (guardanapos + porta)", 30.00, 59.90),
        ("Centro de mesa com bordado inglês", 22.00, 41.90),
        ("Toalha de mesa infantil plastificada", 20.00, 38.90),
        ("Almofada com estampa boho", 18.00, 34.90),
        ("Cortina blackout 2m x 1,80m", 65.00, 119.90),
        ("Tapete sala peludo 1,5m x 2m", 95.00, 179.90),
        ("Manta decorativa para sofá", 40.00, 74.90),
        ("Capa de almofada (kit 4 unidades)", 30.00, 59.90),
        ("Quadro decorativo floral", 25.00, 49.90),
        ("Vaso decorativo pequeno", 15.00, 29.90),
        ("Cortina para cozinha estampada", 35.00, 64.90),
        ("Kit decoração quarto infantil", 45.00, 89.90),
        ("Luminária decorativa de mesa", 55.00, 99.90),
        ("Abajur tecido rústico", 48.00, 94.90),
        ("Cesto de roupas em algodão cru", 38.00, 74.90),
        ("Bandeja decorativa com espelho", 28.00, 54.90),
        ("Relógio de parede floral", 30.00, 58.90),
        ("Porta-chaves decorativo de madeira", 20.00, 39.90),
        ("Painel macramê para parede", 35.00, 69.90),
        ("Porta-joias de cerâmica artesanal", 22.00, 42.90),
        ("Quadro triplo em MDF decorativo", 40.00, 78.90),
    ], start=1)
])

# 2. Generate Sales and Purchases for May, June, July 2025
months = ["2025-05", "2025-06", "2025-07"]
sales_rows = []
purchases_rows = []
for month in months:
    for _, row in products.iterrows():
        qty_sold = np.random.randint(0, 20)
        sales_rows.append({
            "Mês": month,
            "Código": row["Código"],
            "Produto": row["Produto"],
            "Qtde Vendida": qty_sold,
            "Receita (R$)": qty_sold * row["Venda (R$)"]
        })
        qty_buy = np.random.randint(10, 50)
        purchases_rows.append({
            "Mês": month,
            "Código": row["Código"],
            "Produto": row["Produto"],
            "Qtde Comprada": qty_buy,
            "Custo Total (R$)": qty_buy * row["Custo (R$)"]
        })

sales = pd.DataFrame(sales_rows)
purchases = pd.DataFrame(purchases_rows)

# 3. Expenses
expenses = pd.DataFrame([
    {"Mês": month, "Descrição": "Aluguel", "Valor (R$)": 1200.00} for month in months
] + [
    {"Mês": month, "Descrição": "Internet + Energia", "Valor (R$)": 400 + np.random.randint(-20, 21)} for month in months
] + [
    {"Mês": month, "Descrição": "Marketing", "Valor (R$)": 200 + np.random.randint(-50, 51)} for month in months
])

# 4. Cashflow Summary
summary = []
for month in months:
    total_sales = sales[sales["Mês"] == month]["Receita (R$)"].sum()
    total_purchases = purchases[purchases["Mês"] == month]["Custo Total (R$)"].sum()
    total_expenses = expenses[expenses["Mês"] == month]["Valor (R$)"].sum()
    summary.append({
        "Mês": month,
        "Vendas (R$)": total_sales,
        "Compras (R$)": total_purchases,
        "Despesas (R$)": total_expenses,
        "Saldo (R$)": total_sales - total_purchases - total_expenses
    })
cashflow = pd.DataFrame(summary)

# 5. Export to Excel
file_path = '/mnt/data/emporio_grande_ouro_dados.xlsx'
with pd.ExcelWriter(file_path) as writer:
    products.to_excel(writer, sheet_name='Produtos', index=False)
    sales.to_excel(writer, sheet_name='Vendas', index=False)
    purchases.to_excel(writer, sheet_name='Compras', index=False)
    expenses.to_excel(writer, sheet_name='Despesas', index=False)
    cashflow.to_excel(writer, sheet_name='Cashflow', index=False)

# 6. Dashboard Charts
# Chart 1: Monthly Revenue vs Cashflow
plt.figure()
plt.plot(cashflow["Mês"], cashflow["Vendas (R$)"], marker='o')
plt.plot(cashflow["Mês"], cashflow["Saldo (R$)"], marker='o')
plt.title("Vendas e Saldo Mensal")
plt.xlabel("Mês")
plt.ylabel("Valor (R$)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Chart 2: Top 10 Produtos por Qtde Vendida (Período Total)
total_by_product = sales.groupby("Produto")["Qtde Vendida"].sum().sort_values(ascending=False).head(10)
plt.figure()
plt.bar(total_by_product.index, total_by_product.values)
plt.title("Top 10 Produtos por Quantidade Vendida")
plt.xlabel("Produto")
plt.ylabel("Qtde Vendida")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

file_path
```
